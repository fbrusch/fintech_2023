```python
def create_birthdays(n):
    
    return [randint(1,365) for _ in range(n)]
```

```python
create_birthdays(100)
```

```python
sum([are_there_duplicates(create_birthdays(26)) for _ in range(1000)])/1000
```

```python
def estimate_probability_of_collision(n):
    return sum([are_there_duplicates(create_birthdays(n)) for _ in range(1000)])/1000
    
```

```python
estimate_probability_of_collision(27)
```

```python
probs = [estimate_probability_of_collision(x) for x in range(10,51)]
```

```python
fs
```

```python
from math import prod
```

```python
prod(fs)
```

```python
def compute_probability_of_collision(n):
    
    factors = [(365-m)/365 for m in range(1,n)]
    return 1-prod(factors)
```

```python
compute_probability_of_collision(10)
```
