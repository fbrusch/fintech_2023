```python
from random import random
```

```python
def throw_stone():
    x = random()*2-1
    y = random()*2-1
    return x, y
```

```python
throws = [throw_stone() for _ in range(100)]
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
for t in throws:
    plt.plot(t[0],t[1],"ob")
```

```python
throws[10]
```

```python
def is_in_circle(t):
    return (t[0]**2+t[1]**2) < 1
```

```python
is_in_circle(throws[10])
```

```python
for t in throws:
    if is_in_circle(t):
        plt.plot(t[0],t[1],"or")
    else:
        plt.plot(t[0],t[1],"ob")
```

```python
throws_within_the_circle = [t for t in throws if is_in_circle(t)]
```

```python
len(throws_within_the_circle)
```

```python
def estimate_pi(n):
    return sum([1 for _ in range(n) if ((random()*2-1)**2+(random()*2-1)**2) < 1])/n*4
```

```python
estimate_pi(10**6)
```

```python
estimates = [estimate_pi(2**n) for n in range(20)]
```

```python
plt.plot(estimates)
```
